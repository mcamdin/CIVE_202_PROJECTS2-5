# Define the IDM Function

```python
def idm_acceleration(v, s, delta_v, v0, s0, T, a, b, delta):
    
    s_star = s0 + max(0, v * T + (v * delta_v) / (2 * np.sqrt(a * b)))
    
    accel = a * (1 - (v / v0)**delta - (s_star / s)**2)
    
    return accel
```

# Prepare the Data for the IDM Simulation

```python
trajectory_number = 1

data_subset = ngsim_data[ngsim_data['trajectory_number'] == trajectory_number].reset_index(drop=True)

time_data = data_subset['Time'].values
leader_position = data_subset['leader_position(m)'].values
leader_speed = data_subset['leader_speed(m/s)'].values

dt = time_data[1] - time_data[0]
```

```python
v0 = 30
s0 = 2
T = 1.5
a_param = 1.0
b = 1.5
delta = 4

n_steps = len(time_data)

sim_position = np.zeros(n_steps)
sim_speed = np.zeros(n_steps)
sim_acceleration = np.zeros(n_steps)

sim_position[0] = data_subset['follower_position(m)'].iloc[0]
sim_speed[0] = data_subset['follower_speed(m/s)'].iloc[0]
```

# Run the IDM Simulation

```python
for i in range(n_steps - 1):
    
    gap = leader_position[i] - sim_position[i]
    
    if gap <= 0:
        gap = 0.1
    
    delta_v = sim_speed[i] - leader_speed[i]
    
    sim_acceleration[i] = idm_acceleration(sim_speed[i], gap, delta_v, v0, s0, T, a_param, b, delta)
    
    sim_speed[i + 1] = sim_speed[i] + sim_acceleration[i] * dt
    
    if sim_speed[i + 1] < 0:
        sim_speed[i + 1] = 0
    
    sim_position[i + 1] = sim_position[i] + sim_speed[i] * dt
```

# Plot Actual vs Simulated Follower Position

```python
plt.figure(figsize=(12,5))

plt.plot(time_data, data_subset['follower_position(m)'], label='Actual Follower Position', color='black', linewidth=2)
plt.plot(time_data, sim_position, label='Simulated Follower Position', color='red', linewidth=2)

plt.title('Actual vs Simulated Follower Position', fontsize=14)
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('Position (m)', fontsize=12)

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
